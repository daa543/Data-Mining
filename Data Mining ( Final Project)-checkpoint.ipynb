# Example 1 (K-means)

```python
from IPython.display import Image, display
```

```python
display(Image(filename="C:/Users/Doaa Samir/Pictures/Screenshots/Screenshot 2023-08-26 131446.png"))
```

### Code

```python
import pandas as pd
import numpy as np

# visualiation
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("C:/Users/Doaa Samir/Downloads/kmeans1.xlsx")
df
```

```python
# display the values in array
x=df.values
x
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k,init="k-means++")
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11),wcss,marker='8')
plt.xlabel("K value")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans=KMeans(n_clusters=3)
label=kmeans.fit_predict(x)
print(label)
```

```python
print(kmeans.cluster_centers_)
```

```python
plt.scatter(x[:,0],x[:,1],c=kmeans.labels_,cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color='black')
plt.title('Clusters of Customers')
plt.xlabel('Annual income')
plt.ylabel('Spending Score(1-100)')
plt.show()
```

### Initialialize three clusters

```python
initial_centroids= np.array([[2, 6], [5,10],[6,11]])
```

```python
k = 3 # Change this to your desired number of clusters

# Initialize an empty list to store the WCSS values
wcss = []

# Fit K-Means with the specified number of clusters and custom initial centroids
kmeans = KMeans(n_clusters=k, init=initial_centroids)
kmeans.fit(x)
wcss.append(kmeans.inertia_)
```

```python
label=kmeans.fit_predict(x)
print(label)
```

```python
print(kmeans.cluster_centers_)
```

```python
plt.scatter(x[:,0],x[:,1],c=kmeans.labels_,cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color='black')
plt.title('Clusters')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### Is this new entry will be red or blue?

```python
display(Image(filename="C:/Users/Doaa Samir/Pictures/Screenshots/Screenshot 2023-08-24 101549.png"))
```

### Code

```python
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_excel("C:/Users/Doaa Samir/Downloads/knn1.xlsx")
df
```

```python
x=df['Brightness']
y=df['Saturation']
classes=df['Class']
```

```python
scatter = plt.scatter(x,y,c=classes)
```

```python
class_values = list(set(classes))
```

```python
class_labels = ['Class {}'. format(c) for c in class_values]
```

```python
plt.legend(scatter.legend_elements()[0], class_labels)
plt.show()
```

```python
data = list(zip(x,y))
data
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(data,classes)
```

```python
new_x = 20
new_y = 35
new_point = [(new_x, new_y)]
```

```python
prediction = knn.predict(new_point)
print(prediction)  # 1 means red class
```

```python
plt.scatter(x + [new_x], y + [new_y], c=classes + [prediction[0]])
plt.text(x=new_x-1.7, y=new_y-0.7, s=f"new point, class: {prediction[0]}")
plt.xlim(10, 100)
plt.ylim(10, 150)
plt.show()
```

###  updated scatter plot code to show the new  point

```python
# Convert the x, y, and classes to pandas Series
x = pd.Series(x)
y = pd.Series(y)
classes = pd.Series(classes)
```

```python
# Concatenate the new point's coordinates and class to the existing Series
x = pd.concat([x, pd.Series([new_x])])
y = pd.concat([y, pd.Series([new_y])])
classes = pd.concat([classes, pd.Series([prediction[0]])])
```

```python
plt.scatter(x, y, c=classes)
plt.text(x=new_x-1.7, y=new_y-5.7, s=f"new point, class: {prediction[0]}")
plt.xlim(5, 100)
plt.ylim(5, 130)
plt.show()
```
